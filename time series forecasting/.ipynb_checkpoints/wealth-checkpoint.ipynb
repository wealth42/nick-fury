### Splitting the dataset for model building

```python
#train and test splitting
train = ds[0:int(len(ds)*0.7)] 
test = ds[int(len(ds)*0.7):]
```

```python
train.count()
```

```python
test.count()
```

```python
# extract out the time-series
feature_train =  train['bike_count']
feature_test =  test['bike_count']
```

# Statistics

```python
#(Augmented Dickey Fuller) test
dftest = adfuller(feature_train)
dftest
print('DF test statistic is %3.3f' %dftest[0])
print('DF test p-value is %1.4f' %dftest[1])
```

## Observation
#### p-value <= 0.05: Reject the null hypothesis (H0), the data does not have a unit root and is stationary. ,

```python
feature_ts_train_diff = feature_train.diff(periods=1)
feature_ts_train_diff.dropna(inplace=True)

feature_ts_test_diff = feature_test.diff(periods=1)
feature_ts_test_diff.dropna(inplace=True)
```

```python
fig, axes = plt.subplots(1, 2)
fig.set_figwidth(12)
fig.set_figheight(4)
plt.xticks(range(0,30,1), rotation = 90)
smt.graphics.plot_acf(feature_ts_train_diff, lags=30, ax=axes[0])
smt.graphics.plot_pacf(feature_ts_train_diff, lags=30, ax=axes[1])
plt.tight_layout()
```

## Observation
##### we can see the negative correlation with partial  ,  variables are changing in the opposite direction , whilst the autocorrelation somewhat okay 


# Forescasting the data

## Autoregressive Integrated Moving Average (ARIMA) Model


```python
# Build Model 
model = ARIMA(feature_ts_train_diff, order=(0, 1, 1)) 
fitted = model.fit()  

# Forecast
fc, se, conf = fitted.forecast(147, alpha=0.05)  

# Make as pandas series
fc_series = pd.Series(fc, index=test.index)
lower_series = pd.Series(conf[:, 0], index=test.index)
upper_series = pd.Series(conf[:, 1], index=test.index)

# Plot
plt.figure(figsize=(8,5), dpi=100)
plt.plot(feature_ts_train_diff, label='training')
plt.plot(feature_ts_test_diff, label='actual')
plt.plot(fc_series, label='forecast')
plt.fill_between(lower_series.index, lower_series, upper_series, 
                 color='k', alpha=.15)
plt.ylim(-70,50)
plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

### Observation
Chosen the combination of p,q,d after trying other combinations , but still it is not improving 

# Holt-Winters

```python
y_hat_avg = test.copy()
fit1 = ExponentialSmoothing(np.asarray(train['bike_count']) ,seasonal_periods=7 ,trend='add', seasonal='add',).fit()
y_hat_avg['Holt_Winter'] = fit1.forecast(len(test))
plt.figure(figsize=(16,8))
plt.plot( train['bike_count'], label='Train')
plt.plot(test['bike_count'], label='Test')
plt.plot(y_hat_avg['Holt_Winter'], label='Holt_Winter Prediction')
plt.legend(loc='best')
plt.ylim(-1,50)
plt.show()
```

#### Observation

 The prediction is not bad , it can be better with high volume of data

### Error rate

```python
rms = sqrt(mean_squared_error(test.bike_count, y_hat_avg.Holt_Winter))
print(rms)
```

######  Error rate is not to high not too low but it is okay for this dataset

# Conclusion

##### The data is stationary. Where p is equal to the zero , so it may change on the large volume of future dataset.
##### Partial Correlation is worst , so the data of ACF(yt and yt–h) does not have that much of  Correlation.
##### For this dataset Holt winter will be a better choice of model for prediction , but even in the future we can go for ARIMA if it is need .
